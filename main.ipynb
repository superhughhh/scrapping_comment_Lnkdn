```python
from bs4 import BeautifulSoup
import re
import pandas as pd

class MyScrapper:
    def __init__(self, folder_path):
        self.folder_path = folder_path
        self.soup = ""

    def file_loader(self):
        path = self.folder_path
        print(path)
        with open(path, 'r', encoding='utf-8') as file:
            html_content = file.read()
        self.soup = BeautifulSoup(html_content, 'html.parser')
```

```python
post_page_path = r"/Users/odekerkenhugues/Desktop/Script/Scrapping_nom_commentaire_Linkedin/page/post_2.html"

a = MyScrapper(post_page_path)
a.file_loader()
#comment_part = a.soup.find(id="ember485")
comment_list = a.soup.findAll(attrs={"class":"comments-comment-item"})
#len(comment_list)
ID = 0
prospect_dict = {}
for comment in comment_list:
    #name = comment.findAll("a")[1].find(class_="comments-post-meta__name-text hoverable-link-text mr1").get_text(strip=True)#.find('span', attrs={"aria-hidden":"true"})#[1].get_text(strip=True)
    try :
        name = comment.findAll("a")[1].find(class_="comments-post-meta__name-text t-14 hoverable-link-text").get_text(strip=True)#pas pour tous AttributeError
    except AttributeError:
        name = comment.findAll("a")[1].find(class_="comments-post-meta__name-text hoverable-link-text mr1").get_text(strip=True)
        name = name.split("Voir")[0]

    job = comment.findAll("a")[1].find(class_="comments-post-meta__headline t-12 t-normal t-black--light").get_text(strip=True)
    comment_body = comment.find(class_="update-components-text relative").get_text(strip=True)
    fiche_prospect= {}
    print(name)
    fiche_prospect["nom"] = name
    print(job)
    fiche_prospect["m√©tier"] = job
    print(comment_body)
    fiche_prospect["commentaire"] = comment_body
    prospect_dict[ID] = fiche_prospect
    ID+=1
```
